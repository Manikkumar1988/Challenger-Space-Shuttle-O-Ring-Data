```python
# Import libraries necessary for this project
import numpy as np
import pandas as pd
from sklearn.cross_validation import ShuffleSplit

# Pretty display for notebooks
%matplotlib inline

# Load the Space Shuttle O-Ring dataset
data = pd.read_csv('o-ring-erosion-only.data', delim_whitespace=True,names=['O_rings', 'distress', 'temperature', 'pressure', 'order'], header=None)
o_ring = data['distress']
features = data.drop('distress', axis = 1)
    
# Success
print "Space Shuttle O-Ring dataset has {} data points with {} variables each.".format(*data.shape)
```

## Feature Observation

As a reminder, we are using four features from the Challenger Space Shuttle O-Ring dataset: 'distress', 'temperature', 'pressure' and 'order'. For each data point :

    'o_ring' Number of O-rings at risk on a given flight 
    'distress' Number experiencing thermal distress 
    'temperature' Launch temperature (degrees F) 
    'pressure'  Leak-check pressure (psi) 
    'order' Temporal order of flight
